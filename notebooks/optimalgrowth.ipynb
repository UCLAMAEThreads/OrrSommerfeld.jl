# Optimal growth of disturbances

```julia
using OrrSommerfeld
using Plots
```

```julia
Re = 5000
α, β = 1.0, 0.0

dOS = os_matrices(α,β,Re,baseflow=:poiseuille);
eOS = os_eigen(dOS,ilims=(-1.5,Inf),normalize=:true);
```

```julia
opt = optimal_growth(dOS,(0.0,150.0));
```

```julia
κ0 = zeros(ComplexF64,length(eOS.values));
κ0[1] = 1.0

E = disturbance_energy(κ0,dOS,eOS,opt.t);
```

```julia
plot(opt.t,opt.G,xlim=(0,Inf),ylim=(0,Inf))
plot!(opt.t,E)
```

```julia
x, y, u = velocity_x(opt.flow0,dOS);
```

```julia
plot(x, y, u, ratio=1)
```

```julia
x, y, u = velocity_x(opt.flowt,dOS);
plot(x, y, u, ratio=1)
```
